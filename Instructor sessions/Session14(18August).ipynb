```python
car_data = pd.read_csv('Cleaned_Car_Data.csv')
car_data
```

```python
car_data.info()
```

```python
X = car_data.drop(columns='Price')
X
```

```python
y = car_data['Price']
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
from sklearn.compose import make_column_transformer
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)
```

```python
oho = OneHotEncoder()
oho.fit(X[['name', 'company','year','fuel_type']])
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories=oho.categories_),['name', 'company','year','fuel_type']), remainder='passthrough')
```

```python
column_trans
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr= LinearRegression()
```

```python
pipe = make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
y_pred
```

```python
predictions = pd.DataFrame({'Original Price':y_test, "Predicted Price": y_pred})
predictions
```

```python
r2_score(y_test, y_pred)
```

```python
scores=[]
for i in range(1000):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
```
