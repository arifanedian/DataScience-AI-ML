```python
def convert(obj):
    
    L =[]
    for i in ast.literal_eval(obj):             
        L.append(i['name'])
    return L      
```

```python
movies.genres = movies.genres.apply(convert)
```

```python
def convert_cast(obj):
    l=[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=3:
            l.append(i['name'])
            counter +=1
        else:
            break
        
    return l
```

```python
#top 3 cast
movies['cast'] = movies['cast'].apply(convert_cast)
movies['cast']
```

### write a function that returns the name of the director of the film from the crew table

```python
def get_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break # there will be only one director in a movie.
        
    return l
```

```python
movies['crew']= movies['crew'].apply(get_director)
movies
```

# Vectorization

```python
import nltk
from nltk.corpus import stopwords
stopwords.words('english')
```

# Stemming

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
new_df['tags'][0]
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
new_df['tags']
```

```python
# Bag of Words 
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000, stop_words='english',binary=True)
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
vectors.shape
```

```python
# cv.get_feature_names()
```

# Similarity Check

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
similarity.shape
```

```python
similarity
```

```python
# need to get top 5 high sililarity scores
# we can sort it in descending order

sorted(similarity[0],reverse=True)


#This will sort the series but index postion will also be moved
# we will not only sort but also generate new indices usinf enumerate
```

```python
# this will sort on new indices 
# sorted(list(enumerate(similarity[0])))


#this will held the indexes also 
sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x: x[1])[1:10]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse=True, key=lambda x: x[1])[1:6]
    for i in movie_list:
        print(new_df.iloc[i[0]]['title'])
```

```python
recommend('Avatar')
```

```python
recommend("The Dark Knight Rises")
```

```python
recommend("Pirates of the Caribbean: At World's End")
```
