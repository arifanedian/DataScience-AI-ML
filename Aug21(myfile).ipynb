# Predicting Car Price

```python
cardata = pd.read_csv('data files/Cleaned_Car_Data.csv')
```

```python
cardata.year
```

```python
cardata.info()
```

```python
cardata = cardata[cardata['Price']<5000000]
```

```python
cardata.info()
```

# Extracting Training Data

```python
X = cardata[['name','company','year','kms_driven','fuel_type']]
y = cardata['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer
from sklearn.metrics import r2_score
```

### Creating a one hot encoder object to contain all possible categories

```python
ohe = OneHotEncoder()
ohe.fit(X[['name','company','year','fuel_type']])
```

### Creating a Column transformer to transform categorical columns

```python
column_transform = make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','year','fuel_type']),remainder='passthrough')
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_transform,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
predictions = pipe.predict(X_test)
```

```python
r2_score(y_test,predictions)
```

```python
scores=[]
for i in range(1000):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_transform,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
```
